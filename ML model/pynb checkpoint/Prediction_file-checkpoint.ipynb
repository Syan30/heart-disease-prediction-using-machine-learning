# Model Implementation

```python
Input = Heart_Data.drop('target',axis=1)
Input
```

```python
Output = Heart_Data['target']
Output
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(Input, Output, test_size = 0.2, random_state=0)
```

# Naive Bayes Algorithm

```python
from sklearn.naive_bayes import GaussianNB
Model_NB = GaussianNB()
Model_NB.fit(X_train, y_train)
Predition_NB = Model_NB.predict(X_test)
```

```python
Predition_NB
```

```python
from sklearn.metrics import accuracy_score
Accuracy_NB = accuracy_score(y_test,Predition_NB)
print('Accuracy of Naive Bayes Algorithm',Accuracy_NB)
```

# Classification Report of Naive Bayes Algorithm

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, Predition_NB))
```

# Confusion Matrix  of Naive Bayes Algorithm

```python
from sklearn.metrics import confusion_matrix

CM = confusion_matrix(y_test, Predition_NB)
```

```python
# visualize confusion matrix with seaborn heatmap

cm_matrix = pd.DataFrame(data=CM, columns=['Actual Positive:1', 'Actual Negative:0'], 
                                 index=['Predict Positive:1', 'Predict Negative:0'])

sns.heatmap(cm_matrix, annot=True, fmt='d', cmap='YlGnBu')
```
